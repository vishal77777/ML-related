```python
# replacing a value 
dfss=df.copy(deep=True)
dfss.head()

dfss["Sex"]=dfss["Sex"].replace(["female","male"],["woman","man"])
dfss["Sex"]
```

```python
#renaming columns
dfss=dfss.rename(columns={'PClass':'Passenger class'})
dfss.head()
dfss=dfss.rename(columns={'PClass':'Passenger class','Sex':'Gender'})
dfss.head()
```

```python
#handling missing values
dfss[dfss['Age'].isnull()].head(2)
```

```python
import numpy as np

#to use full functionality of NAN
dfss['Gender']=dfss['Gender'].replace('man',np.nan)
dfss[dfss['Gender'] == 'NaN']
```

```python
#some operations using a column
dfss[dfss['Gender'].isnull()]
dfss['Gender']=dfss['Gender'].replace(np.nan,'man')


count=dfss['Gender'].count()-1
for i in range(0,5):
 print(dfss['Name'][count],dfss['Gender'][count])
 count-=1

dfss.tail()

 

```

```python
#dropping columns and rows
dfss.head(15)
dfss.drop(0,axis=0)
dfss.drop('Age',axis=1)
dfss.head(5)
```

```python
#dropping duplicates
dfss.drop_duplicates()#drop by rows
dfss.drop_duplicates(subset=['Gender'],keep='first') #drop by specific colmns
dfss.drop_duplicates(subset=['Gender'],keep='last')
```
